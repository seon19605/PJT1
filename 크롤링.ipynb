```python
from selenium import webdriver
driver = webdriver.Chrome()
url = "http://search.danawa.com/dsearch.php?query=무선청소기&tab=main"
driver.get(url)
```

```python
#html = 웹사이트의 구성 중 하나
from bs4 import BeautifulSoup

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
prod_items = soup.select("li.prod_item") 
len(prod_items)
#prod_items = 하나만 가져오는거임
#0번 1번 2번으로 그 정보만 가져오는거임
```

```python
title = prod_items[0].select("p.prod_name > a")[0].text.strip()

print(title)
```

```python
spec_list = prod_items[0].select("div,spec_list")[0].text.replace("\t","").replace("\n","")
print(spec_list)
```

```python
price = prod_items[0].select("li:nth-child(1) > p.price_sect > a > strong")[0].text.strip().replace(",","")
print(price)
```

```python
def get_prod_items(prod_items):
    prod_data = []
    for prod_item in prod_items:
        if 'product-pot' in prod_item['class']:
            continue
        try:
            #1.샘플링 가져오기
            title = prod_item.select('p.prod_name > a')[0].text.strip()
            #2. 스팩목록 가져오기
            spec_list = prod_item.select('div.spec_list')[0].text.replace('\t', '').replace('\n', '') #\는 엔터키 위에꺼
            #3. 가격정보 가져오기
            price = prod_item.select('li:nth-child(1) > p.price_sect > a > strong')[0].text.strip().replace(",", "")
            prod_data.append([title, spec_list, price])
        except:
            pass
    return prod_data
```

```python
prod_items = soup.select('div.main_prodlist > ul.product_list > li.prod_item')
print(prod_items)
prod_data = get_prod_items(prod_items)
print(len(prod_data))
print(prod_data)
```

```python
def get_search_page_url(keyword, page):
    return 'http://search.danawa.com/dsearch.php?query={}&volumeType=allvs&page={}&limit=30&sort=saveDESC&list=list&boost=true&addDelivery=N&tab=goods&tab=goods'.format(keyword, page)

keyword = '아이패드'
page = 1
url = get_search_page_url(keyword, page) #위에 링크의{}= keyword
print(url)
```

```python
import time

keyword = '무선청소기'
total_page = 10
prod_data_total = [] #모든 데이터 저장
driver = webdriver.Chrome()

for page in range(1, total_page + 1): #1부터 토탈페이지까지
    #검색페이지로 이동
    url = get_search_page_url(keyword, page)
    driver.get(url)
    
    #페이지가 로딩 될때까지 대기
    time.sleep(5)
    
    #현재 페이지의 html(사이트를 이루는 코드들) 정보 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    #상품 정보 추출
    prod_items = soup.select('div.main_prodlist > ul.product_list > li.prod_item')
    prod_items_list = get_prod_items(prod_items)
    
    #추출 데이터 저장
    prod_data_total = prod_data_total + prod_items_list #각 페이지별로 뽑힌 아이템리스트
```

```python
#표형식으로 정리하기
import pandas as pd
data = pd.DataFrame(prod_data_total)
data.columns = ['상품명', '스펙 목록', '가격']
data.to_excel('./1_danawn_crawling_result.xlsx', index = False)
#크롤링 파일에 있는거는 엑셀(앱)에서 열기
```
